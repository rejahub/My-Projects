Apakah YrSold berpengaruh terhadap harga jual dan Setiap tahunnya harga rumah mengalami kenaikan?

# Penentuan Hipotesis

Uji Hipotesis sisi kiri

H0 : harga rumah <= harga tahun sebelumnya

H1 : harga rumah > harga tahun sebelumnya

α = 5% = 0.05

```python
y2006 = ndf[ndf['YrSold']==2006]
y2007 = ndf[ndf['YrSold']==2007]
y2008 = ndf[ndf['YrSold']==2008]
y2009 = ndf[ndf['YrSold']==2009]
y2010 = ndf[ndf['YrSold']==2010]
```

```python
anova_test = st.f_oneway(y2006['SalePrice'],y2007['SalePrice'],y2008['SalePrice'],y2009['SalePrice'],y2010['SalePrice'])
```

```python
anova_test.pvalue
```

```python
if anova_test.pvalue > 0.05:
    print('Ho diterma')
else:
    print('Ho ditolak')
```

Kesimpulan : Ho diterima, sehingga harga rumah tidak selalu naik setiap tahunnya.
