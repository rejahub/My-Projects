```python
class DataPreparation:
  def __init__(self,dataTraffic):
    self.dataTraffic = dataTraffic

  def ResultTraffic(self):
    resultTrafficDict = {}
    resultTrafficList = []
    for data in dataTraffic:
      for device in data['device']:
        resultTrafficDict['date'] = data['date']
        resultTrafficDict['device'] = list(device.keys())[0]
        resultTrafficDict['visit'] = list(device.values())[0]
        resultTrafficList.append(resultTrafficDict.copy())
    return(resultTrafficList)

  def AgregasiTraffic(self):
    agregasiTrafficDict = {}
    agregasiTrafficList = []
    for data in dataTraffic:
      totalTraffic = 0
      for device in data['device']:
        traffic = list(device.values())[0]
        totalTraffic += traffic
      agregasiTrafficDict['date'] = data['date']
      agregasiTrafficDict['traffic'] = totalTraffic
      agregasiTrafficList.append(agregasiTrafficDict.copy())
    return(agregasiTrafficList)


dataPreparation = DataPreparation(dataTraffic)
resultTraffic = dataPreparation.ResultTraffic()
print('result traffic ={}'.format(resultTraffic))
agregasiTraffic = dataPreparation.AgregasiTraffic()
print('agregasi traffic = {}'.format(agregasiTraffic))
```
